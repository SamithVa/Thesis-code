```python
# Percentage Graph

import matplotlib.pyplot as plt
import numpy as np

# Define x-axis thresholds from 0.5 to 0.9 with an interval of 0.1.
thresholds = np.arange(0.5, 1.0, 0.1)

# Data for each resolution: the values represent the number of similar tokens at each threshold.
# The total number of tokens for each resolution is given by the numeric label.
data_1440 = [1406, 1377, 1301, 1048, 452]  # Total tokens: 1440 (Resolution: 1344x840)
data_1000 = [966, 923, 845, 640, 228]       # Total tokens: 1000 (Resolution: 1120x700)
data_660  = [623, 591, 524, 342, 83]          # Total tokens: 660  (Resolution: 924x560)
data_476  = [444, 412, 342, 200, 45]          # Total tokens: 476  (Resolution: 784x476)

# Compute percentage of similar tokens relative to the total token count for each resolution.
perc_1440 = [value / 1440 * 100 for value in data_1440]
perc_1000 = [value / 1000 * 100 for value in data_1000]
perc_660  = [value / 660 * 100 for value in data_660]
perc_476  = [value / 476 * 100 for value in data_476]

# Initialize a figure with a size appropriate for academic presentations.
plt.figure(figsize=(10, 6))

# Plot the percentage data for each resolution using distinct markers and consistent line widths.
plt.plot(thresholds, perc_1440, marker='o', linewidth=2, label='1440 tokens (1344x840)')
plt.plot(thresholds, perc_1000, marker='s', linewidth=2, label='1000 tokens (1120x700)')
plt.plot(thresholds, perc_660,  marker='^', linewidth=2, label='660 tokens (924x560)')
plt.plot(thresholds, perc_476,  marker='d', linewidth=2, label='476 tokens (784x476)')

# Set the axis labels with increased font sizes for improved clarity.
plt.xlabel('Cosine Similarity', fontsize=12)
plt.ylabel('Percentage of Similar Visual Tokens (%)', fontsize=12)

# Provide an informative and descriptive title.
plt.title('Percentage of Similar Visual Tokens vs Cosine Similarity', fontsize=14)

# Explicitly specify x-axis tick marks for the defined thresholds.
plt.xticks(np.arange(0.5, 1.0, 0.1))

# Set the y-axis limit to a full percentage range for clarity.
plt.ylim(0, 100)

# Add a legend with a title that denotes the resolution and token total.
plt.legend(title='Visual Tokens (Resolution)', fontsize=10, title_fontsize=12)

# Enable and configure major and minor grid lines for detailed reference.
plt.grid(which='major', linestyle='--', linewidth=0.7, color='gray')
plt.minorticks_on()
plt.grid(which='minor', linestyle=':', linewidth=0.5, color='lightgray')

# Adjust the layout to ensure no elements are clipped.
plt.tight_layout()

# Display the resulting plot.
plt.show()
```
